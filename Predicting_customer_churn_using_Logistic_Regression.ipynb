```python
# keeping important columns 
keep_to_columns=['gender','SeniorCitizen','Partner','Dependents','tenure','PhoneService','MultipleLines','Contract','TotalCharges','Churn']
df=df[keep_to_columns]
```

```python
# encoded binary variables
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
```

```python
categorical_cols=['gender','Partner','Dependents','PhoneService','MultipleLines','Contract','Churn']
for col in categorical_cols:
    df[col]=label_encoder.fit_transform(df[col])
```

```python
# split independent and dependent data from dataset(df)
x=df.drop('Churn',axis=1)
y=df['Churn']
```

```python
# train_test_split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
# here total charges is a obejct type but we need to convert the numeric value using(to_numeric)
x_train['TotalCharges']=pd.to_numeric(df['TotalCharges'],errors='coerce')
x_test['TotalCharges']=pd.to_numeric(df['TotalCharges'],errors='coerce')
```

```python
# check x_train is null or not (is null then fillna with their mean value)
x_train.isnull().sum()
```

```python
# fillna
x_train['TotalCharges'].fillna(x_train['TotalCharges'].mean(),inplace=True)
```

```python
# check x_test is null or not (is null then fillna with their mean value)
x_test.isnull().sum()
```

```python
# fillna
x_test['TotalCharges'].fillna(x_test['TotalCharges'].mean(),inplace=True)
```

```python
# standardization process(optional but offten benifical for Logistic Regression)
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
x_train=scaler.fit_transform(x_train)
x_test=scaler.fit_transform(x_test)
```

```python
## logistic Regression
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
# accuracy
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

# ## classification system

```python
## classification system
def predictive(gender,SeniorCitizen,Partner,Dependents,tenure,PhoneService,MultipleLines,Contract,TotalCharges):
    data={
        'gender':[gender],
        'SeniorCitizen':[SeniorCitizen],
        'Partner':[Partner],
        'Dependents':[Dependents],
        'tenure':[tenure],
        'PhoneService':[PhoneService],
        'MultipleLines':[MultipleLines],
        'Contract':[Contract],
        'TotalCharges':[TotalCharges]
    }
    df1=pd.DataFrame(data)
    
  

    # Encode the categorical columns
    categorical_columns = ['gender', 'SeniorCitizen', 'Partner', 'Dependents', 'PhoneService', 'MultipleLines', 'Contract']
    for column in categorical_columns:
        df1[column] = label_encoder.fit_transform(df1[column])
    df1 = scaler.fit_transform(df1)

    result = lr.predict(df1).reshape(1,-1)
    return result[0]

```

```python
gender='Female'
SeniorCitizen='No'
Partner='Yes'
Dependents='No'
tenure=1
PhoneService='No'
MultipleLines='No phone service'
Contract='Month-to-month'
TotalCharges=29.85
result=predictive(gender,SeniorCitizen,Partner,Dependents,tenure,PhoneService,MultipleLines,Contract,TotalCharges)

if result==0:
    print('Not Churn')
else:
    print('Churn')
```
